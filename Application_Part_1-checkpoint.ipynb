Import libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from sklearn import preprocessing
from matplotlib import pyplot as plt
from scipy.spatial.distance import pdist, squareform
```

Data loading

```python
MARKET_PRICES=pd.read_excel('APPLICATION_PART_1.xlsx', 'MARKET_PRICES',usecols='B:BU')
NAMES=list(MARKET_PRICES)
MARKET_PRICES=np.array(MARKET_PRICES)
TR_PRICES=np.array(pd.read_excel('APPLICATION_PART_1.xlsx', 'INDICES_REINVESTED DIVIDENDS',usecols='B:BU'))
n_stocks=MARKET_PRICES.shape[1]
n_obs=MARKET_PRICES.shape[0]
```

1.1. Euclidean distances

1.1.a. Rescaling returns: log of the total return prices rebased on their average level

```python
SCALED_PRICES=np.log(TR_PRICES/np.mean(TR_PRICES,axis=0))
```

1.1.b. Distance matrix

```python
DIST_MATRIX=squareform(pdist(SCALED_PRICES.T,'euclidean'))
```

1.1.c. Identification of the 5 pairs that display the smallest distance, in the PAIRS_DIST(5x3) vector.

    1st col - distance
    2nd col - ID stock A (long leg of the spread)
    2nd col - ID stock B (short leg of the spread)

```python
dist_vector=np.zeros((int(n_stocks*(n_stocks-1)/2),3))
k=0
for i in range(n_stocks):
    for j in np.arange(i+1,n_stocks,1):
        dist_vector[k,0]=DIST_MATRIX[i,j]
        dist_vector[k,1]=i
        dist_vector[k,2]=j
        k=k+1
PAIRS_DIST=dist_vector[(dist_vector[:,0]).argsort()][:5,:]
```

1.1.d. Chart of the rescaled stock prices of the selected pairs

```python
for i in range(5):
    print(NAMES[int(PAIRS_DIST[i,1])]+'-'+NAMES[int(PAIRS_DIST[i,2])])
    plt.plot(SCALED_PRICES[:,PAIRS_DIST[i,1:].astype(int)])
    plt.show()
```

1.2. Similarity based on the correlation matrix

1.2.a. Daily returns

```python
RETURNS=TR_PRICES[1:,:]/TR_PRICES[:-1,:]-1
```

1.2.b. Correlation Matrix

```python
CORR_MATRIX=np.corrcoef(RETURNS,rowvar=0)
```

1.2.c. Identifying the pairs that display the highest return correlation in the PAIRS_CORR(5x3) vector.

    1st col - correlation coefficient
    2nd col - ID stock A (long leg of the spread)
    2nd col - ID stock B (short leg of the spread)

```python
corr_vector=np.zeros((int(n_stocks*(n_stocks-1)/2),3))
k=0
for i in range(n_stocks):
    for j in np.arange(i+1,n_stocks,1):
        corr_vector[k,0]=CORR_MATRIX[i,j]
        corr_vector[k,1]=i
        corr_vector[k,2]=j
        k=k+1
PAIRS_CORR=corr_vector[(-corr_vector[:,0]).argsort()][:5,:]
```

1.2.d. Chart of the rescaled stock prices of the selected pairs

```python
for i in range(5):
    print(NAMES[int(PAIRS_CORR[i,1])]+'-'+NAMES[int(PAIRS_CORR[i,2])])
    plt.plot(SCALED_PRICES[:,PAIRS_CORR[i,1:].astype(int)])
    plt.show()
```

2.1. Spread computation

```python
SPREADS=np.zeros((n_obs,5))
for i in range(5):
    SPREADS[:,i]=SCALED_PRICES[:,int(PAIRS_DIST[i,1])] - SCALED_PRICES[:,int(PAIRS_DIST[i,2])]
```

2.2. Charts of the spreads

```python
for i in range(5):
    plt.plot(SPREADS[:,i])
    plt.show()
```

3.1. Stationarity test for low distance pairs

Perform an ADF test. AR(1) model with a constant, and only 1 lag

Record the p-values in the AD_PVALUES vector

```python
AD_PVALUES=np.zeros((5))
for i in range(5):
    AD_PVALUES[i]=adfuller(SPREADS[:,i], maxlag=1,regression='c',)[1]
    if AD_PVALUES[i]<0.05:
        print('Spread '+NAMES[int(PAIRS_DIST[i,1])]+'-'+NAMES[int(PAIRS_DIST[i,2])]+' -> Stationary')
    else:
        print('Spread '+NAMES[int(PAIRS_DIST[i,1])]+'-'+NAMES[int(PAIRS_DIST[i,2])]+' -> Non-stationary')
```

4.1. Spread normalization 

```python
NORM_SPREADS=preprocessing.scale(SPREADS)
```

4.2. Computation of the number of zero-crossings

Tip: count one crossing each time the sign of the spread changes between two consecutive dates

```python
print(np.sum((NORM_SPREADS[1:,:]/NORM_SPREADS[:-1,:])<0,axis=0))
```
